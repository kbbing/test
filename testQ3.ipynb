```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('coffee.csv')

total_orders = df.groupby(['product_category', 'product_type']).size().reset_index(name='total_order')
total_orders = total_orders.sort_values(by='product_type')

total_orders.to_csv('coffee_total_order.csv', index=False)

plt.figure(figsize=(12, 6))
plt.barh(total_orders['product_type'], total_orders['total_order'])
plt.title('Total Order for each Product Type')
plt.ylabel('product_type')
plt.xlabel('total_order')
plt.tight_layout()
plt.show()
```
